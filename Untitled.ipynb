然后点开文件夹中saved_resource.html就是你的文件地址

```python
from bs4 import BeautifulSoup

# 打开本地HTML文件
with open('D:/Desktop/saved_resource.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
```

```python
# 使用BeautifulSoup解析HTML
soup = BeautifulSoup(html_content, 'html.parser')

# 找到HTML中的表格元素
table = soup.find_all('table')
```

```python
row_data = []
for item in table:
    for column in item.tbody.find_all("tr"):
        col_dict = {}
        atoms = column.find_all("td")
        count = 0
        temp_dict = {}
        for atom in atoms:
            count += 1
            if count == 1:
                temp_dict["course_name"] = atom.get_text()
            if count == 4:
                temp_dict["credit"] = atom.get_text()
            if count == 5:
                temp_dict["grade_point"] = atom.get_text()
            if count == 6:
                if atom.get_text() == "优":
                    temp_dict["result"] = 90
                elif atom.get_text() == "良":
                    temp_dict["result"] = 80
                elif atom.get_text() == "中":
                    temp_dict["result"] = 70
                elif atom.get_text() == "及格":
                    temp_dict["result"] = 60  
                elif atom.get_text() == "免修":
                    temp_dict["result"] = 80   
                else:
                    temp_dict["result"] = atom.get_text()
        row_data.append(temp_dict)
            
```

```python
total_score = 0  # 初始化总分
total_credit = 0  # 初始化总学分
avg_score = 0
for course in row_data:
    credit = float(course['credit'])
    grade_point = float(course['grade_point'])
    weighted_score = credit * grade_point  # 计算加权分数GPA
    total_score += weighted_score  # 累加到总分
    total_credit += credit  # 累加到总学分
    
for course in row_data:
    credit = float(course['credit'])
    grade_point = float(course['result'])
    weighted_score = credit * grade_point / total_credit  # 计算加权分数
    avg_score += weighted_score  # 累加到总分
    

weighted_average_gpa = total_score / total_credit  # 计算加权平均分
print("加权GPA平均分: {:.2f}".format(weighted_average_gpa))
print("加权平均分: {:.2f}".format(avg_score))
```
